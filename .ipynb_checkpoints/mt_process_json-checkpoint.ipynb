```python
EVENT_SELLING = 0


ADJ = 'adj '
SYMBOL = 'sym '
DATE_EVENT = 'dev '
ACTION_SHORTING = 'sel '
ACTION_SHORTED = 'sold'
MONTH = 'mnt '
STRIKE_DATE = 'sd  '
STRIKES = 'stk '
SPREAD_TYPE = 'spt '
PRICE = 'pri '
PRICE_PER_SHARE = 'pps '
STOP = 'stp '
CONDITION = 'cnd '
WATCH = 'wat '
COVER = 'cvr '
CONDOR = 'cndr'
    
class Adj():
    pass

class Symbol():
    sym = None
    def __init__(self,sym):
        self.sym = sym
    def __repr__(self):
        return self.sym
    
s = ''
o = []
m = []

def seq_push(pos,letter,obj):
    global s,o
    
    s = s + letter
    o.append(obj)
    
    obj.index = pos
    
def det_symbol(text):
        
    if any(char.isdigit() for char in text):
        return False
        
    if text.upper() != text:
        return False
    
    if text.startswith('DTE'):
        return False
            
    if text.startswith('ADJ'):
        return False
            
    if text == 'I':
        return False
            
    if text.isalpha() is False:
        return False
            
    return True

class Obj_base():
    valid = False
    index = None

class Slash_pair(Obj_base):
    left = None
    right = None
    def __init__(self,text,form):

        s = text.split('/')
        if len(s)!=2:
            return
        
        try:
            self.left = form(s[0])
        except:
            return
        
        try:
            self.right = form(s[1])
        except:
            return
        
        self.valid = True
        
    def __repr__(self):
        return str(self.left) + '/' + str(self.right)

    
class Month_slash_day(Obj_base):
    sp = None
    
    def __init__(self,text):
        
        text1 = text.rstrip(':')
        if text1 == text:
            return
        
        self.sp = Slash_pair(text1,int)
        if not self.sp.valid:
            return
        
        if self.sp.left > 12:
            return
        if self.sp.right > 31:
            return
        
    
        self.valid = True
        
    def __repr__(self):
        return str(self.sp) + ':'
        
class Action_shorting(Obj_base):
    shorting = None
    
    def __init__(self,text):
    
        self.valid = True
    
        if text.lower() == 'shorting':
            self.shorting = True
            return
            
        if text.lower() == 'selling':
            self.shorting = True
            return
        
        self.valid = False
        
    def __repr__(self):
        return "shorting(" + str(self.shorting) + ")"

        
class Action_shorted(Obj_base):
    shorted = None
    
    def __init__(self,text):
    
        self.valid = True
    
        if text.lower() == 'shorted':
            self.shorted = True
            return
            
        if text.lower() == 'sold':
            self.shorted = True
            return
        
        self.valid = False
        
    def __repr__(self):
        return "shorted(" + str(self.shorted) + ")"

        
class Month(Obj_base):
    month_num = None
    def __init__(self,text):
        
            for i,m in enumerate(['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']):
                if text.lower() == m:
                    if i==4:
                        if text[0] == 'm':  # May must have capitol 'M'
                            return
                    self.month_num = i
                    self.valid = True
                    return
                
    def __repr__(self):
        return 'Month[' + str(self.month_num) + ']'
                
class Strikes(Obj_base):
    sp = None
    def __init__(self,text):
#        if text[0]!='$':
#            return
        text = text.lstrip('$')
        text = text.rstrip('.,')
        text = text.replace('bull','')  # incase of typo of no space after strikes
        text = text.replace('bear','')
        self.sp = Slash_pair(text,float)
        self.valid = self.sp.valid
        
    def __repr__(self):
        if self.sp is None:
            return '(None/None)'
        return '(' + str(self.sp.left) + '/' + str(self.sp.right) + ')'
        
class Strike_date(Obj_base):
    year = None
    month = None
    day = None
    
    def process_dte(self,timestamp,last):

        try:
            num = int(last)
        except:
            return
        
        ts = dateutil.parser.isoparse(timestamp)
        end_date = ts + datetime.timedelta(days=num)
        
        self.year = end_date.year
        self.month = end_date.month
        self.day = end_date.day
    
        self.valid = True

        
    def __init__(self,text,timestamp,last):
        if text.lower()=='dte':
            self.process_dte(timestamp,last)
        else:
            if text[0] != '(':
                return
            text = text.lstrip('(')
            text.rstrip('.,:')
            if text[-1] != ')':
                return
            text = text.rstrip(')')
            sp = Slash_pair(text,int)
            if sp.valid:
                #print("Strike_date",timestamp)
                ts = dateutil.parser.isoparse(timestamp) # ISO 8601 exteended form
                ts2 = dateutil.parser.isoparse(timestamp)
                ts2 = ts2.replace(month=sp.left,day=sp.right)
                if ts2 < ts:
                    ts2 = ts2.replace(year = int(ts2.year) + 1)
                self.year = ts2.year
                self.month = ts2.month
                self.day = ts2.day
        
                self.valid = sp.valid
            
    def __repr__(self):
        return str(self.month) + '/' + str(self.day) + '/' + str(self.year)
                
        
class Spread_type(Obj_base):
    put = None
    def __init__(self,text):
        if text.lower()=='put':
            self.put = True
        if text.lower()=='call':
            self.put = False
            
        if self.put != None:
            self.valid = True
            
    def __repr__(self):
        if self.put is True: return 'PUT'
        if self.put is False: return 'CALL'
        return 'None'
            
    
            
class Price(Obj_base):
    price = None
    def __init__(self,text):
#        if text[0]!='$':
#            return
#        text = text.lstrip('$')
#        text = text.rstrip('.,)')
        
        non_decimal = re.compile(r'[^\d.]+')
        text = non_decimal.sub('',text)
        text = text.rstrip('.')
        try:
            self.price = float(text)
            self.valid = True
        except:
            pass
        
    def __repr__(self):
        if self.price is not None:
            return '$' + str(self.price)
        return 'None'
        
class Price_per_share(Obj_base):
    price = None
    def __init__(self,text):
        sp = text.split('/')
        if len(sp)!=2:
            return
        if sp[1].lower().rstrip(',.') != 'share':
            return
        p = Price(sp[0])
        self.valid = p.valid
        self.price = p.price
        
    def __repr__(self):
        if self.price is not None:
            return '$' + str(self.price) + '/share'
        else:
            return 'None'
        
class Stop(Obj_base):

    def __init__(self,text):
        text = text.lstrip('(')
        if text.lower() == 'stop':
            self.valid = True
    def __repr__(self):
        if self.valid:
            return 'Stop'
        else:
            return 'None'
            
class Condition(Obj_base):
    over = None
    def __init__(self,text):
        if text.lower() == 'over':
            self.over = True
        if text.lower() == 'under':
            self.over = False
            
        self.valid = (self.over != None)
        

            
class Watch(Obj_base):
    def __init__(self,text):
        if text.lower().startswith('watch'):
            self.valid = True
            
class Cover(Obj_base):
    def __init__(self,text):
        if text.lower().startswith('cover'):
            self.valid = True
            
class Condor(Obj_base):
    def __init__(self,text):
        if text.lower().startswith('condor'):
            self.valid = True
        
def det_month(text):month = Month(text);return month.valid;
def det_date_event(text):msd = Month_slash_day(text);return msd.valid
def det_shorting(text):action = Action_shorting(text);return action.valid
def det_shorted(text): action = Action_shorted(text);return action.valid
def det_strike_date(text,timestamp,last):strike_date = Strike_date(text,timestamp,last);return strike_date.valid
def det_strikes(text):strikes = Strikes(text);return strikes.valid
def det_spread_type(text):spread_type = Spread_type(text);return spread_type.valid
def det_price(text):price = Price(text);return price.valid
def det_price_per_share(text):pps = Price_per_share(text);return pps.valid
def det_stop(text): stop = Stop(text);return stop.valid
def det_condition(text): condition = Condition(text);return condition.valid
def det_watch(text): watch = Watch(text);return watch.valid
def det_cover(text): cover = Cover(text);return cover.valid
def det_condor(text): condor = Condor(text);return condor.valid



def filter(symb,text):
    
    text_split = text.split()
    
    for i,w in enumerate(text_split):
        
        if det_symbol(w):
            #print("filter",symb,w, w == symb,text)
            return w == symb
            
    return False

    
    
def create_tokens(timestamp,text):
    
    global s,o
    
#    print('create_tokens:',text)
    
    s = ''
    o = []
    
    text_split = text.split()
    
    last = None

    for i,w in enumerate(text_split):
        
        if i:
            last = text_split[i-1]
                
        if w.startswith('ADJ'):
            seq_push(i,ADJ,Adj())
            continue
            
        if det_symbol(w):
            seq_push(i,SYMBOL,Symbol(w))
            continue
            
        if det_date_event(w):
            seq_push(i,DATE_EVENT,Month_slash_day(w))
            continue
            
        if det_shorting(w):
            seq_push(i,ACTION_SHORTING,Action_shorting(w))
            continue
            
        if det_shorted(w):
            seq_push(i,ACTION_SHORTED,Action_shorted(w))
            continue

        
            
        if det_month(w):
            seq_push(i,MONTH,Month(w))
            continue
            
        if det_strike_date(w,timestamp,last):
            seq_push(i,STRIKE_DATE,Strike_date(w,timestamp,last))
            continue
            
        if det_strikes(w):
            seq_push(i,STRIKES,Strikes(w))
            continue
            
        if det_spread_type(w):
            seq_push(i,SPREAD_TYPE,Spread_type(w))
            continue
            
        if det_stop(w):
            seq_push(i,STOP,Stop(w))
            continue
            
        if det_price(w):
            seq_push(i,PRICE,Price(w))
            continue
            
        if det_price_per_share(w):
            seq_push(i,PRICE_PER_SHARE,Price_per_share(w))
            continue
            
        if det_condition(w):
            seq_push(i,CONDITION,Condition(w))
            continue
            
        if det_watch(w):
            seq_push(i,WATCH,Watch(w))
            continue
            
        if det_cover(w):
            seq_push(i,COVER,Cover(w))
            continue
            
        if det_condor(w):
            seq_push(i,CONDOR,Condor(w))

{'timestamp': '2020-05-29T05:13:30', 'event': 0, 'sell': True, 'exp_month': 5, 'exp_day': 29, 'strike_low': 180, 'strike_high': 185, 'put': False}

def has_class(cl):
    for obj in o:
        if type(obj) is cl:
            return True
    return False

def get_obj(cl):
    for obj in o:
        if type(obj) is cl:
            return obj
        
    return None

def get_obj_after(cl1,cl2):
    
    found = False
    for obj in o:
        if found is False:
            if inspect.isclass(cl1):
                if type(obj) is cl1:
                    found = True
            else:
                if obj is cl1:
                    found = True
        else:
            if type(obj) is cl2:
                return obj
    return None
    

count = 0

def count_shorted(timestamp,text):
    
    shorted1 = get_obj(Action_shorting)
    shorted2 = get_obj_after(shorted1,Action_shorting)
    
    if shorted1 and shorted2:
        print(text)
    pass

def get_stop(timestamp,text):
    global o
    
    text_split = text.split()

    symbol = get_obj(Symbol)
    
    if not symbol:
        print('get stop: no symbol',text)
        return
    
    stop = get_obj(Stop)
    
    
    if stop:
        if len(text_split)-1 > stop.index:
            next = text_split[stop.index + 1]
            price = Price(next)
            if price.valid:
                pass
                #print(symbol,'stop1:',price,next)
            else:
                print(symbol,'stop word:',next)
    else:
        print(symbol,"no stop")

    

def get_options(timestamp,text):
    global o
    
    options = []
    
    action = None
    symbol = None
    
    for obj in o:
        
        if not symbol:
            if type(obj) is Symbol:
                symbol = obj

        if action:
            if type(obj) is Spread_type:
                spread_type = obj
            if type(obj) is Strike_date:
                strike_date = obj
            if type(obj) is Strikes:
                strikes = obj
                
            if symbol and spread_type and strike_date and strikes:
                option = {}
                option['action'] = action
                option['sym'] = symbol
                option['spread_type'] = spread_type
                option['strike_date'] = strike_date
                option['strikes'] = strikes
                options.append(option)
                
                action = None

        else:
            if type(obj) is Action_shorting:
                action = obj
            if type(obj) is Action_shorted:
                action = obj
                
#            print('action',str(action))
                
            spread_type = strike_date = strikes = None
            
    if len(options):
        print(options)
    else:
        print(text)
    
            
    return options
            
            

def process_text(timestamp,text):
    global count
    
#    if filter('SPX',text) is False:
#        return
    
#    print(text)
    
    create_tokens(timestamp,text)
    
    get_stop(timestamp,text)
    
    #options = get_options(timestamp,text)
    #count_shorted(timestamp,text); return
    return


    mrk = {}
    mrk['timestamp'] = timestamp
    
    condition = get_obj(Condition)

    symbol = get_obj(Symbol)
    shorting = get_obj(Action_shorting)
    spread_type = get_obj_after(selling,Spread_type)
    strike_date = get_obj_after(selling,Strike_date)
    strikes = get_obj_after(selling,Strikes)
    stop = get_obj_after(selling,Stop)
    stop_price = get_obj_after(stop,Price)
    if not stop_price:
        stop_price = get_obj_after(stop,Price_per_share)
    condor = get_obj(Condor)

    
    #if not condition and selling:
    if shorting:

        if symbol and spread_type and strike_date and strikes and stop_price:
            count  = count + 1
            mrk['event'] = EVENT_SELLING
            mrk['sell'] = True
            
            print('found complete:',str(symbol), str(spread_type),str(strike_date),str(strikes),str(stop_price))
            #print(text)
    
        else:
            pass
            print('found selling incomplete',str(spread_type),str(strike_date),str(strikes),stop_price,text)
        
    else:
        if symbol and spread_type and strike_date and strikes and stop_price:
            print('move stop: ',str(symbol), str(spread_type),str(strike_date),str(strikes),str(stop_price))
        pass
        #print(text)
    

    
def process_element(timestamp,element):


    if isinstance(element,str):
        process_text(timestamp,element)
        
    if isinstance(element,list):
        
        #remove the embedded dictionaries and concat the strings
        
        complete = ''
        
        for l in element:
            if isinstance(l,str) is not True:
                continue
                
            complete = complete + l
            
        process_text(timestamp,complete)


print('channel id:',data['id'])
print("spread_type","strike_date","strikes")
for message in data['messages']:
#    if message['type'] != 'message':
#        continue

#    print(message)
    
    if 'text' in message.keys():
        process_element(message['date'],message['text'])

#    try:
#        text = message['text']
#        process_element('date',text)
#    except:
#        print('failure:',message)
#        break
        
print(count)
        
    
        

```
