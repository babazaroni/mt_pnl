```python
def get_underlying_bars(qb,sym,start_time,end_time):
    equity = qb.AddEquity(sym)
    history = qb.History(equity.Symbol, start_time, end_time, Resolution.Minute)
    print("underlying:",sym,len(history))
    #print(history.head(2))
    #print(history.tail(2))
    return history

def filter_sym(sym):
    if sym in ['HES','AHC','EA','ERTS','UAL','UAUA','MKC','MCCRK','VXX','VXXB','WFC','NOB',
                       'AZN','ZEN','NKLA','VTIQ','TGT','DH']:
        #print("skipping bad:", event['sym'])
        #pass
        return False
    
    if sym in ['ZM','PYPL','AAPL','GS','UAL','LYFT','WYNN','GOOGL','RAD','GRUB','BA','W','MA',
                       'QQQ','MRNA','SPY','EA','NVDA','SPY','SPX','MAR','AMD','COST','SBUX','DKNG',
                       'NFLX','SMH','BYND','COST','FB','V','LULU','BILL','RCL','HD','LOW','TWTR',
                       'WORK','AMZN','ROKU','KR','XOP','DDOG','DXCM','ZS','CRM','TSLA','SPCA','TEAM',
                       'PENN','MSFT','JBLU','COUP','DAL','MCD','CVS','LLY','BURL','PLAY','SQ','DIA',
                       'RH','NCLH','APT','FSLY','CIEN','DIS','NKE','PTON','OSTK','NCLH','MU','SPCE',
                       'INTC','IYT','XLF','IWM','KMX','NEM','NOC','SWK','YUM','UBER','USO','GLD','WDAY',
                        'ACB','xXLI','BLMN','DOCU','TWLO','YETI','ADBE','WMT','MS','XLE','CODX','TDOC',
                       'LVS','DELL','SHOP','MTCH','ANET','CAT','AAL','TNA','LUV','GD','DRI']:
        #print("skipping good:", event['sym'])
        #pass
        return False
    
    return True

    
    
```

```python
def get_option_bars(qb,sym,ts,strike_date,spread_type,strikes):
    #equity = qb.AddEquity("SPY", Resolution.Minute)
    print("\rchecking:",sym)
    
    equity = qb.AddEquity(sym, Resolution.Minute)
    
    data = []
    lengths = []
    
    contracts = qb.OptionChainProvider.GetOptionContractList(equity.Symbol,ts)
    #print("contracts: ",len(contracts))
    for strike in strikes:
        for c in contracts:
            if c.ID.OptionRight ==  (0 if spread_type == 'CALL' else 1):
                if strike == c.ID.StrikePrice and c.ID.Date.date() == strike_date.date():
                    #print(event['sym'],c.ID.Symbol,c.ID.OptionRight,c.ID.Date.date(),c.ID.StrikePrice,c.ID.Market,c.ID.SecurityType)
                    option_history = qb.GetOptionHistory(c, ts,strike_date)
                    d = option_history.GetAllData()
                    data.append(d)
                    lengths.append(len(d))
                    
                    #print("------------- HEAD ---------------")
                    #print(d.head(2))
                    #print("------------- TAIL ---------------")
                    #print(d.tail(2))
                    
                    break
                    
    if (len(lengths) and (lengths[0] != lengths[-1])):
        print("-------------------------------- Error in lengths: ",lengths)
    return data
```

```python
class Tick_event:
    event = None
    time = None
    sort_key = 0
    priority = 1  # for sorting, price updates 0, events 1

    
    def __init__(self,event):
        self.event = event
        self.time = datetime.strptime(event['ts'], '%Y-%m-%dT%H:%M:%S') + timedelta(hours=3)
        self.sort_key = self.time.timestamp()*10 + self.priority
    
    def on_tick(self):
        event = self.event
        
        if all(item in event.keys() for item in ['action','sym','spread_type','strike_date','strikes']):
        
            #if filter_sym(event['sym']) is False:
            #    continue
            
            exp = datetime.strptime(event['strike_date'], '%Y-%m-%d') + timedelta(hours=16)
            if self.event['action'] == 'shorting':
            

                end_time = datetime.strptime(event['strike_date'], '%Y-%m-%d').replace(hour=16, minute = 0, second = 0)
            
                self.underlying_bars = get_underlying_bars(qb,
                            event['sym'],
                            self.time,
                            end_time)
            
                self.option_bars = get_option_bars(qb,
                           event['sym'],
                           self.time,
                           end_time,
                           event['spread_type'],
                           event['strikes'])
            
                #print('{} shorting: {} {} {} {}'.format(ts,event['sym'],event['spread_type'],exp,event['strikes']))
            if 'stop' in event.keys():
                #print('{} stop: {}'.format(start_time,event['stop']))
                pass

        return []
    
    def __repr__(self):
        return str(self.event)
```

```python
tick_buf = []

for event in actions:
    tick_buf.append(Tick_event(event))
    
#print("before sort",tick_buf)
    
tick_buf.sort(key=lambda x: x.sort_key)

#print("after sort",tick_buf)

while (len(tick_buf)):
    tick = tick_buf.pop(0)
    new_ticks = tick.on_tick()
    if len(new_ticks):
        tick_buf.extend(new_ticks)
        tick_buf.sort(key=lambda x: x.sort_key)
    
    
```
